## Перевод на русский

Используем ChatGPT для перевода статей

```python
!mkdir ru_md
!pip install -q openai
```

```python
from openai import OpenAI
from google.colab import userdata

# gets API Key from environment variable OPENAI_API_KEY
client = OpenAI(
  base_url="https://openrouter.ai/api/v1",
  api_key=userdata.get('OpenAIToken'),
)
```

```python
from concurrent.futures import ThreadPoolExecutor
from tqdm import tqdm
from functools import partial


def call_openai_api(entry,pbar):
    answer = client.chat.completions.create(
        model="openai/gpt-3.5-turbo",
        messages=[
            {
            "role": "user",
            "content": f"Translate markdown from English to Russian. Preserve equations form :\n {entry}"
            },
        ],
    )
    pbar.update(1)
    return answer



def process_markdown(array):
    with tqdm(total=len(array)) as pbar:
        with ThreadPoolExecutor() as executor:
            carry_pbar = lambda x: call_openai_api(x,pbar)
            responses = list(executor.map(carry_pbar, array))
        return [response.choices[0].message.content for response in responses]
```

```python
from pathlib import Path

output = []

for path in tqdm(Path('mmd_output').iterdir()):
    output.append(process_markdown(path.read_text().split('\n\n')))
```

```python
for i,path in enumerate(Path('mmd_output').iterdir()):
    with open(f'ru_md/ru_{path.name}','w') as f:
        f.write('\n'.join(output[i]))
```

```python
en_ver = []

for i,path in enumerate(Path('mmd_output').iterdir()):
    en_ver.append(path.read_text())
```

```python
import pandas as pd

df = pd.DataFrame({
    'en':en_ver,
    'ru': ['\n'.join(output[i]) for i in range(len(output))]
    }
)
```
